```python
import pandas as pd

df_train_en = pd.read_csv('/Users/patriciagrigor/Desktop/MASTER/CEU_hate/multilingual-hate-speech/Data/IMSyPP_EN YouTube/IMSyPP_EN_YouTube_comments_train.csv')

df_train_slo = pd.read_csv('/Users/patriciagrigor/Desktop/MASTER/CEU_hate/multilingual-hate-speech/Data/IMSyPP_SI Twitter/IMSyPP_SI_anotacije_round1.csv')

df_train_it = pd.read_csv('/Users/patriciagrigor/Desktop/MASTER/CEU_hate/multilingual-hate-speech/Data/IMSyPP_IT YouTube/IMSyPP_IT_YouTube_comments_round1.csv')
```

```python
type_counts = df_train_en['Type'].value_counts()
type_counts
```

```python
#remove type "0" from english dataset
df_train_en = df_train_en[df_train_en.Type != '0']
```

```python
#show counts for each class in the English training set
df_train_en['Type'].value_counts()
```

```python
import matplotlib.pyplot as plt

# Counting the number of comments for each label
type_counts = df_train_en['Type'].value_counts()

# Define the order of the labels
labels_order = ['0. appropriate', '1. inappropriate', '2. offensive', '3. violent']

# Reindex type_counts to ensure the correct order
type_counts = type_counts.reindex(labels_order)

# Creating the bar plot
plt.figure(figsize=(10, 4))
plt.bar(type_counts.index, type_counts.values, color='palevioletred')
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Label distribution for the English dataset')
plt.show()
```
